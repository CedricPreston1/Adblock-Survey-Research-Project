```python
df = df.rename(columns= {'Current_users_of_ad_blockers' : 'Current users of ad blockers', 'Past_users_of_ad_blockers' : 'Past users of ad blockers'})
df #renaming the last two columns
```

Fixing the answers of "Current users of ad blockers" and "Past users of ad blockers"

```python
df2 = df.copy()
```

```python
df2
```

```python
df2[['Current users of ad blockers', 'Past users of ad blockers']] = df2[['Current users of ad blockers', 'Past users of ad blockers']].replace({1 : 'Yes', 0 : 'Other'})
df2 #this is what I want to rename it to for these two columns, "Other" because the answer could have been anything else that wasn't "No"
```

Now I will need to rename the scale answers from Numerical to worded answers. This is what it looks like 

3 Completely Agree

2 Agree

1 Somewhat Agree

0 Neutral 

-1 Somewhat Disagree

-2 Disagree

-3 Completely Disagree

The source from kaggle only refers 3 and -3 as "Completely Agree' and "Completely Disagree" Respectuflly

So i'm just going to assum -2 and 2 are "Highly disagree" and "Highly agree" and so on.

```python
df3 = df2.copy()
df3 #Another copy just in case
```

```python
replace_map = {-3 : 'Completely Disagree',
               -2 : 'Disagree',
               -1 : 'Somewhat Disagree',
               0 : 'Neutral',
               1 : 'Somewhat Agree',
               2 : 'Agree',
               3 : 'Completely Agree'} 
```

```python
df3 = df3.replace(replace_map)
df3
```

Looking at some of the data

```python
import matplotlib.pyplot as plt
```

```python
Current_users = df3['Current users of ad blockers'].value_counts()
Past_users = df3['Past users of ad blockers'].value_counts() #setup
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,8), sharey=True)

axes[0].bar(Current_users.index, Current_users.values, color='deepskyblue')
axes[0].set_title('Current users of Ad Blockers')
axes[0].set_xlabel('Response')
axes[0].set_ylabel('Count')

axes[1].bar(Past_users.index, Past_users.values, color= 'limegreen')
axes[1].set_title('Past users of Ad Blockers')
axes[1].set_xlabel('Response')

plt.tight_layout()
plt.show() #For the overall
```

```python
#Here I'm lowering the age range down to 35 and under and filtering the sex to male to see if there's a factor here.
filtered_df = df3[(df3['Age'] <= 35) & (df3['Sex'].str.lower()=='male')]
grouped_current = filtered_df.groupby('Sex')['Current users of ad blockers'].value_counts().unstack().fillna(0)
grouped_past = filtered_df.groupby('Sex')['Past users of ad blockers'].value_counts().unstack().fillna(0)

fig, axes = plt.subplots(1, 2, figsize=(12,8), sharey=True)

grouped_current.plot(kind='bar', ax=axes[0], color=['crimson', 'deepskyblue'])
axes[0].set_title('Current users of Ad Blockers')
axes[0].set_xlabel('Response')
axes[0].set_ylabel('Count')

grouped_past.plot(kind='bar', ax=axes[1], color=['crimson', 'limegreen'])
axes[1].set_title('Past users of Ad Blockers')
axes[1].set_xlabel('Response')

plt.tight_layout()
plt.show()
```
